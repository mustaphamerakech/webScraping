```python
df.drop('extracted_information', axis=1, inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



# Liste des colonnes catégorielles à encoder
categorical_columns = ['company_name', 'job_name', 'localisation', 'work_shift', 'experience']

# Instancier LabelEncoder
label_encoder = LabelEncoder()

# Appliquer LabelEncoder à chaque colonne catégorielle
for col in categorical_columns:
    df[col + '_encoded'] = label_encoder.fit_transform(df[col])

# Maintenant, vous avez des colonnes supplémentaires avec des valeurs encodées
df.drop(['company_name', 'job_name', 'localisation', 'work_shift', 'experience'], axis=1, inplace=True)
df.head()
```

```python
# Exemple d'encodage one-hot pour les colonnes catégoriques
categorical_columns = ['company_name','localisation','work_shift','experience']
df_encoded = pd.get_dummies(df, columns=categorical_columns)
# Afficher le DataFrame résultant
df_encoded.head()
```

```python
X = df_encoded.drop('job_name', axis=1)
X
```

```python
Y = df_encoded.job_name
Y
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
r2
```
